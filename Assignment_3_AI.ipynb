```python
import numpy as np
import matplotlib.pyplot as plt

# Values of n to iterate over
n_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

for n in n_values:
    # Simulate throwing two dice n times and compute the sum
    dice_1 = np.random.randint(1, 7, size=n)
    dice_2 = np.random.randint(1, 7, size=n)
    sums = dice_1 + dice_2

    # Compute frequencies using numpy's histogram function
    h, h2 = np.histogram(sums, bins=np.arange(2, 14))

    # Plot the histogram
    plt.bar(h2[:-1], h / n)
    plt.title(f'Histogram for n = {n}')
    plt.xlabel('Sum of Dice')
    plt.ylabel('Probability')
    plt.show()
```

5 Observations: As n, the number of trials, increases, I will notice that the shape of the histogram becomes more symmetrical and approaches the theoretical probabilities given for each sum. Initially, with fewer trials, you may observe more fluctuations or deviations from the expected probabilities. However, as the number of trials increases, the observed frequencies tend to stabilize and converge towards the expected probabilities.

6 Regression to the Mean: This phenomenon is related to the concept of "regression to the mean." In this context, if we consider extreme outcomes (like getting a sum of 2 or 12 with very few trials), with more trials, these extreme outcomes become less likely, and the observed outcomes tend to regress (move closer) towards the mean outcome (like sums around 7). In other words, extreme outcomes are less sustainable over a large number of trials, leading to a convergence towards the mean. This is evident in the histogram as the number of trials (n) increases.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('weight-height.csv')

plt.figure(figsize=(8, 6))
plt.scatter(data['Height'], data['Weight'], alpha=0.5)
plt.title('Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.grid(True)
plt.show()

X = data[['Height']]
y = data['Weight']

model = LinearRegression()
model.fit(X, y)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.5)
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.title('Linear Regression: Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.grid(True)
plt.show()

y_pred = model.predict(X)

rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print("RMSE:", rmse)
print("R2 Score:", r2)
```
